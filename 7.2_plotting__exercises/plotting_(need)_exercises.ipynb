## Exercise 1

```python
from matplotlib import pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors') #o,x,*,s etc works for marker, markerfacecolor is colour of marker, ecolour is colour of error bar lines
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right') # bottom right does not work
plt.show()
```

## Exercise 2

```python
import numpy as np
```

```python
x = np.linspace(0, 50, 500) #start,stop,number of evenly spaced samples to generate
y = np.exp(-x/10) * np.sin(x)

plt.plot(x, y)
plt.xlabel('x',color='darkorange',fontsize=15)
plt.ylabel('y',color='darkorange',fontsize=15,rotation=0)
plt.title(r'$y=e^{-x/10}\sin(x)$', fontsize=20) #is used to create a raw string in Python. Raw strings treat backslashes (\) as literal characters and prevent them from being interpreted as escape sequences
plt.grid(alpha=0.25)
plt.ylim([-1,1])
plt.show()
```
